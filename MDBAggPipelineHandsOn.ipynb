```python
# 🔧 Import Libraries
from pymongo import MongoClient
import json
```

```python
# 🔌 Connect to MongoDB
client = MongoClient("mongodb+srv://samples:kbdevday@testing.2atn6.mongodb.net/")
db = client['sample_mflix']
collection = db['movies']
```

```python
# 🔍 Run Query 1

for document in collection.find(
    {"rated": 'R'}
  ).limit(5):
    print(json.dumps(document, indent=4, default=str))
```

```python
# 🔍 Run Query 2

for document in collection.find(
    {"year": {"$gt": 2000}}
  ).limit(5):
    print(json.dumps(document, indent=4, default=str))
```

```python
# 🔍 Run Query 3

for document in collection.find(
    {"cast": "Keanu Reeves"}
  ).limit(5):
    print(json.dumps(document, indent=4, default=str))
```

```python
# 🔍 Run Query 4

for document in collection.find(
    {"genres": {'$in': ["Adventure", "Comedy"]}}
  ).limit(20):
    print(json.dumps(document, indent=4, default=str))
```

```python
# 🔍 Run Query 4 with projection

for document in collection.find(
    {"genres": {'$in': ["Adventure", "Comedy"]}},
    {"title": 1, "genres": 1, "_id": 0}
  ).limit(20):
    print(json.dumps(document, indent=4, default=str))
```

```python
# 🔍 Run Aggregation Pipeline 1

pipeline = [
  {
    "$unwind": "$genres"
  },
  {
    "$group": {
      "_id": "null",
      "distinctGenres": {
        "$addToSet": "$genres"
      }
    }
  },
  {
    "$project": {
      "_id": 0,
      "distinctGenres": 1
    }
  }
]

for document in collection.aggregate(pipeline):
    print(json.dumps(document, indent=4, default=str))
```

```python
# 🔍 Run Aggregation Pipeline 1

pipeline = [
  {
    "$unwind": "$genres"
  },
  {
    "$group": {
      "_id": "$genres",
      "movies": {
        "$push": "$title"
      }
    }
  },
  {
    "$project": {
      "genre": "$_id",
      "movies": {
          "$slice": ["$movies", 5]
          },
      "_id": 0 }
  }
]

for document in collection.aggregate(pipeline):
    print(json.dumps(document, indent=4, default=str))
```
